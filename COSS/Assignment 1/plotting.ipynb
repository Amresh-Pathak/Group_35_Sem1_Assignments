```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

__read_clipboard() will read table from clipboard. Hit copy from excel, and run this function.__

```python
direct_map_df = pd.read_clipboard()
direct_map_df
```

```python
plt.figure(figsize=(10,5))
plt.title('Direct Mapping')
p1 = sns.lineplot(data=direct_map_df, y='%Hit Ratio', x='Block Size', hue='Cache size').get_figure().savefig("./plots/direct_mapped.png")
```
